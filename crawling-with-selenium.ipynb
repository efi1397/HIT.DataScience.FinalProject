```python
import time
import bs4
from bs4 import BeautifulSoup  
import pandas as pd
import scipy as sc
import numpy as np
import requests

# --- selenium libraries
import os
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
```

```python
def get_infinity_page_html_data(chromedriver_path="./chromedriver.exe", scroll_number=1, sleep_time=2):
    # Create selenium driver
    s = Service(chromedriver_path)
    driver = webdriver.Chrome(service=s)    
    driver.get(url)
    
    for i in range(1,scroll_number):
        print(f"scroll_number: {i}")
        driver.execute_script("window.scrollTo(1,50000)")
        time.sleep(sleep_time)
    html_data = driver.page_source
    driver.close()
    
    return html_data
    
```

```python
def extract_medium_articles_from_HTML_data(html_data):
    # Extract articles from HTML data 
    soup = BeautifulSoup(html_data, 'html.parser')    
    articles = soup.find_all("article")
    print(f"Number of articles: {len(articles)}")
    
    return articles
```

```python
def create_df_from_articles(articles,category):    
    # Create a lists for df
    categories = []
    headers = []
    subheaders= []
        
    for article in articles:            
        categories.append(category)        
        headers.append(article.find("h2").get_text())
        subheaders.append(article.find("p").get_text())    
    
    #TODO: Append link list
    return pd.DataFrame(
        {"category": categories,  
         "header": headers, 
         "subheader": subheaders})
```

```python
#here you can add more categories
categories = ["technology","politics", "sport"]
    
for category in categories: 
    url = f"https://medium.com/tag/{category}/latest"    
    html_data = get_infinity_page_html_data(scroll_number=4)
    articles = extract_medium_articles_from_HTML_data(html_data)
    df = create_df_from_articles(articles, category)
    print(df.shape)
    print(df)
    print()
```
